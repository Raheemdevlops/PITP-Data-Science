```python
# Student records: name -> {grades, subject}
students = {
    "Alice": {"grades": [85, 92, 78], "subject": "Math"},
    "Bob":   {"grades": [70, 65, 80], "subject": "Science"},
    "Carol": {"grades": [95, 98, 100], "subject": "English"},
}

# Calculate average grade
def average(grades):
    return sum(grades) / len(grades)

# Print report
for name, info in students.items():
    avg = average(info["grades"])
    status = "Pass" if avg >= 75 else "Fail"
    print(f"{name} ({info['subject']}): Avg = {avg:.1f} → {status}")

# Add a new student
students["David"] = {"grades": [60, 72, 68], "subject": "History"}

# Update a student's grades
students["Alice"]["grades"].append(90)

# Find the top student
top = max(students, key=lambda name: average(students[name]["grades"]))
print(f"\nTop student: {top}")
```
